```python
text = """Global carbon emissions from fossil fuels reached record levels again in 2023, as experts warned that the projected rate of warming had not improved over the past two years. The world is on track to have burned more coal, oil and gas in 2023 than it did in 2022, according to a report by the Global Carbon Project, pumping 1.1% more planet-heating carbon dioxide into the atmosphere at a time when emissions must plummet to stop extreme weather from growing more violent.

The finding comes as world leaders meet in Dubai for the fraught Cop28 climate summit. In a separate report published on Tuesday, Climate Action Tracker (CAT) raised its projections slightly for future warming above the estimates it made at a conference in Glasgow two years ago.“Two years after Glasgow, our report is virtually the same,” said Claire Stockwell, an analyst at Climate Analytics and lead author of the CAT report. “You would think the extreme events around the world would be sparking action but governments appear oblivious, somehow thinking treading water will deal with the flood of impacts.”

As carbon clogs the atmosphere, trapping sunlight and baking the planet, the climate is growing more hostile to human life. The growth in CO2 emissions had slowed substantially over the past decade, the Global Carbon Project found, but the amount emitted each year had continued to rise. It projected that total CO2 emissions in 2023 would reach a record high of 40.9 gigatons.If the world continued to emit CO2 at that rate, the international team of more than 120 scientists found, it would burn through the remaining carbon budget for a half-chance of keeping global heating to 1.5C (2.7F) above pre-industrial temperatures in just seven years. In 15 years, the scientists estimated, the budget for 1.7C would be gone too.

The researchers reported big regional differences in emissions. They expected fossil fuel emissions to have risen this year in India and China, the biggest and third-biggest polluters, and to have fallen in the US and the EU, the two biggest historical polluters. The average of the rest of the world’s emissions was expected to have fallen slightly too. Emissions from deforestation and other land-use changes were also projected to have fallen slightly, though not by enough for current levels of tree-planting to make up for it, the researchers found.

For the first time, the scientists also teased out the growth in emissions from flights and ships abroad. The two together were expected to have grown 11.9%, driven by soaring emissions from aviation. Pierre Friedlingstein, a climate scientist at the University of Exeter’s Global Systems Institute and lead author of the study, said: “The impacts of climate change are evident all around us but action to reduce carbon emissions from fossil fuels remains painfully slow. “It now looks inevitable that we will overshoot the 1.5C target of the Paris agreement, and leaders meeting at Cop28 will have to agree rapid cuts in fossil fuel emissions even to keep the 2C target alive.”

Some leaders have also backed efforts to phase out fossil fuels, though only a handful have expressed support for a non-proliferation treaty. Governments were happy to promote clean energy but had done little to penalise fossil fuels, said Glen Peters, a research director at the climate research institute Cicero, who co-wrote the report. “It is simply not enough to support clean energy. Policies are also needed to drive fossil fuels out of the energy system,” he added.

The report also found that technology to suck carbon dioxide out of the atmosphere would have done almost nothing to stop global heating this year. Current levels of technology-based removal – which does not include carbon absorbed by trees – are more than 1m times smaller than current fossil CO2 emissions, the researchers found. Corinne Le Quéré, a research professor at the University of East Anglia’s School of Environmental Sciences, said: “All countries need to decarbonise their economies faster than they are at present to avoid the worst impacts of climate change.”
"""

print(text)
```

```python
# Paraules en minúsucla

lower_text = text.lower()
print (lower_text)
```

Una vegada que hem convertit les paraules a minúscules, procedim a fragmentar-les a través del procés de tokenització.

La **tokenització** implica descompondre un text en fragments més petits (paraules o símbols) amb la finalitat de simplificar el seu examen a l'hora de fer un processament del llenguatge natural. Utilitzarem **NLTK (Natural Language Toolkit)**. L'elecció de posar les paraules en minúscules durant la tokenización és una convenció i té diverses raons: consistència en el tractament de les paraules, reducció de la dimensionalitat, eliminació d'ambigüitat o millora en el rendiment.

```python
# Tokenització

import nltk
nltk.download('punkt')

words = nltk.word_tokenize(lower_text)

print(words[:100])
print(f'\nNombre de paraules al text: {len(words)}')
```

Un cop tokenitzat tot el text, veiem que també estan inclosos els símbols. Ara els eliminarem per tal de facilitar el seu anàlisi més endavant.

```python
from nltk.tokenize import RegexpTokenizer

# Treure els signes de puntuació
tokenizer = RegexpTokenizer(r'\w+')

# Tokenització sense símbols
words = tokenizer.tokenize(lower_text)

print(words[:100])
print(f'\nNombre de paraules al text sense símbols: {len(words)}')
```

Veiem que s'han exclòs els símbols del text. Ara calcularem la freqüència de cada paraula.

```python
from nltk.probability import FreqDist

# Calcular la freqüència de les paraules
freq_dist = FreqDist(words)

freq_dist.plot(30, cumulative=False, linewidth = 3, title='Freqüència de les paraules al text sense símbols')

plt.show()
```

```python
primers_freq_dist = {k: freq_dist[k] for k in list(freq_dist)[:10]}

primers_freq_dist
```

Com era lògic, les paraules amb més presència són les més comunes en la llengua, les anomenades **stopword**. Aquestes paraules són articuladores del llenguatge i no aporten informació significativa, per tant, molt freqüentment, s'obvien en aquests tipus d'anàlisi.

```python
from wordcloud import WordCloud

wordcloud = WordCloud(background_color='white', width=800, height=400).generate_from_frequencies(freq_dist)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Al núvol de paraules també es pot observar com les més freqüents són les stopwords.

### Exercici 2
***
Treu les stopwords i realitza stemming al teu conjunt de dades.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords_eng = set(stopwords.words("english"))
filtered_words = list(filter(lambda word: word not in stopwords_eng, words))

print(filtered_words[:100])
print(f'\nNombre de paraules al text filtrat: {len(filtered_words)}')
```

S'ha reduït considerablement el nombre de paraules després de treure les stopwords.

```python
# Calcular la freqüència de les paraules del text filtrat
freq_dist_filtered = FreqDist(filtered_words)

freq_dist_filtered.plot(30, cumulative=False, linewidth = 3, title='Freqüència de les paraules al text filtrat')
plt.show()
```

```python
primers_freq_dist_filtered = {k: freq_dist_filtered[k] for k in list(freq_dist_filtered)[:10]}

primers_freq_dist_filtered
```

Un cop filtrat, veiem que les paraules que més apareixen al text són paraules relacionades amb la Crisi Climàtica i el medi ambient: emissions, carbon, fossil, climate, etc. Té lògica, ja que és una notícia que parla sobre aquests temes.

```python
from wordcloud import WordCloud

wordcloud_filtered = WordCloud(background_color='white', width=800, height=400).generate_from_frequencies(freq_dist_filtered)

plt.imshow(wordcloud_filtered, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Al núvol de paraules també es pot observar les paraules més freqüents, en aquest cas relacionades amb el Canvi Climàtic.

#### Stemming

El **stemming** és un procés lingüístic que busca reduir les paraules a la seva forma base o arrel, eliminant afixos i sufixos. L'objectiu principal és agrupar paraules relacionades sota una forma comuna. Aquest procés és comunament utilitzat en el processament del llenguatge natural (PLN) per a simplificar l'anàlisi de text.

```python
from nltk.stem import PorterStemmer

stemmed_words = [PorterStemmer().stem(element) for element in filtered_words]

print(stemmed_words[:100])
print(f'\nNombre de paraules al text filtrat amb stemming: {len(stemmed_words)}')
```

En aquest cas no ha variat el nombre de paraules.

### Exercici 3
***
Realitza sentiment analysis al teu conjunt de dades.

El terme **sentiment** es refereix a l'opinió o l'emoció expressada en un text, discurs o una altra forma de comunicació escrita o parlada. En l'àmbit del processament del llenguatge natural (PLN), l'anàlisi de sentiments és una tasca específica que se centra en determinar i comprendre la polaritat emocional darrere d'un fragment de text (si és negatiu, positiu o neutre).

Per realitzar l'anàlisi d'aquest text farem servir **VADER** (Valence Aware Dictionary and sEntiment Reasoner), una llibreria de processament de llenguatge natural dissenyada específicament per a l'anàlisi de sentiment.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

sentiment_text = ' '.join(stemmed_words)

# Realitzar l'anàlisi de sentiment
sentiment = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)

print(sentiment)

# Classificació del sentiment
compound_score = sentiment['compound']

if compound_score >= 0.05:
    sentiment = "Positiu"
elif compound_score <= -0.05:
    sentiment = "Negatiu"
else:
    sentiment = "Neutre"

print(f'\nPuntuació de sentiment: {compound_score}')
print(f'Classificació de sentiment: {sentiment}')
```

En general, la notícia té un component amb tendència negativa. Com hem comentat anteriorment, la notícia parla sobre les emissions i el Canvi Climàtic, per tant, és lògic que sigui negatiu el resultat, ja que no hi ha bones perspectives en aquest àmbit en concret.
